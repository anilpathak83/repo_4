```python
Q1=data.quantile(0.25)
Q3=data.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
# note in linear regression handson we used including condition
# but in logistic regression we r going to use excluding condition
```

```python
data=data[~((data<(Q1-1.5*IQR)) | (data>(Q3+1.5*IQR))).any(axis=1)]
```

# Handeling Missing Values

```python
data.isna().sum()
```

```python
col_list=list(data.columns)
```

```python
col_list
```

```python
col_list.remove('y')
```

```python
col_list
```

```python
for col in col_list:
    if data[col].dtypes=='object':
        data[col]=data[col].fillna(data[col].mode()[0])
    else:
        data[col]=data[col].fillna(data[col].mean())
```
